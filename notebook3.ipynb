```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100)

plt.plot(x, x, label='lineaar')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Three plots")
plt.legend();
plt.show()
```
